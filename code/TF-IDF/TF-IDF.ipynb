```python
import pandas as pd
from docx import Document

# Load the document
doc_path = "/content/drive/MyDrive/clickbait/reports/TFIDF.docx"
doc = Document(doc_path)

# Prepare lists to store data
languages = []
avg_accuracy = []
avg_precision = []
avg_recall = []
avg_f1_score = []
test_accuracy = []
test_precision = []
test_recall = []
test_f1_score = []

# Iterate through paragraphs in the document
for paragraph in doc.paragraphs:
    text = paragraph.text.strip()
    if "Languages:" in text:
        # Split the paragraph into multiple entries if more than one set of data exists
        entries = text.split("Languages:")

        # Iterate over all the split entries to extract data
        for entry in entries[1:]:  # Skip the first empty entry
            try:
                # Extract data after each label and clean extra characters
                langs = entry.split("Average Accuracy:")[0].strip()
                avg_acc = entry.split("Average Accuracy:")[1].split("Average Precision:")[0].strip().replace(',', '')
                avg_prec = entry.split("Average Precision:")[1].split("Average Recall:")[0].strip().replace(',', '')
                avg_rec = entry.split("Average Recall:")[1].split("Average F1 Score:")[0].strip().replace(',', '')
                avg_f1 = entry.split("Average F1 Score:")[1].split("Average Accuracy:")[0].strip().replace(',', '')
                test_acc = entry.split("Test Accuracy:")[1].split("Test Precision:")[0].strip().replace(',', '')
                test_prec = entry.split("Test Precision:")[1].split("Test Recall:")[0].strip().replace(',', '')
                test_rec = entry.split("Test Recall:")[1].split("Test F1 Score:")[0].strip().replace(',', '')
                test_f1 = entry.split("Test F1 Score:")[1].strip().replace(',', '')

                # Append extracted data to corresponding lists
                languages.append(langs)
                avg_accuracy.append((avg_acc))
                avg_precision.append((avg_prec))
                avg_recall.append((avg_rec))
                avg_f1_score.append((avg_f1))
                test_accuracy.append((test_acc))
                test_precision.append((test_prec))
                test_recall.append((test_rec))
                test_f1_score.append((test_f1))

            except (ValueError, IndexError) as e:
                print(f"Error processing entry: {entry} - {e}")
                continue  # Skip to the next entry if an error occurs

# Create a DataFrame with the extracted data
data = {
    "Languages": languages,
    "avg_Accuracy": avg_accuracy,
    "avg_Precision": avg_precision,
    "avg_Recall": avg_recall,
    "avg_F1-Score": avg_f1_score,
    "test_Accuracy": test_accuracy,
    "test_Precision": test_precision,
    "test_Recall": test_recall,
    "test_F1-Score": test_f1_score
}
df = pd.DataFrame(data)

# Save the DataFrame to an Excel file
output_file = "/content/drive/MyDrive/clickbait/reports/TFIDF.xlsx"
df.to_excel(output_file, index=False)

print(f"Data successfully saved to {output_file}")
```

```python
import pandas as pd
import re

# Function to parse data from a text source dynamically
def parse_data_from_text(text):
    data = []

    # Updated pattern to match the format with commas between languages
    pattern = re.compile(
        r"Languages: \((.*?)\), Train Accuracy: ([\d.]+), Test Accuracy: ([\d.]+), Test Precision: ([\d.]+), Test Recall: ([\d.]+), Test F1: ([\d.]+)"
    )

    # Find all matches based on the pattern
    matches = pattern.findall(text)
    print(matches)
    for match in matches:
        print(match)
        languages, train_acc, test_acc, test_prec, test_rec, test_f1 = match
        data.append({
            'Languages': languages,
            'Train Accuracy': float(train_acc),
            'Test Accuracy': float(test_acc),
            'Test Precision': float(test_prec),
            'Test Recall': float(test_rec),
            'Test F1': float(test_f1),
        })

    return data

# Read data from a file dynamically
def read_data_from_file(file_path):
    # Read the text from the file
    print("read")
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Write the data into an Excel file
def write_data_to_excel(data, output_file):
    df = pd.DataFrame(data)
    df.to_excel(output_file, index=False)
    print(f"Data successfully written to {output_file}")

# Main function to read, process, and write data
def main():
    # Specify your input file path (replace 'input.txt' with your actual file path)
    file_path = '/content/drive/MyDrive/clickbait/reports/TFIDF.txt'

    # Read text from the file
    text = read_data_from_file(file_path)
    print(text)
    # Parse data from the text dynamically
    data = parse_data_from_text(text)

    # Specify your output Excel file path
    output_file = 'dynamic_model_results.xlsx'

    # Write the parsed data to the Excel file
    write_data_to_excel(data, output_file)

if __name__ == '__main__':
    main()
```
